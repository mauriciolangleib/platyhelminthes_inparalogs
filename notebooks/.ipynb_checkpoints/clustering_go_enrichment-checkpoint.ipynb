```python
# defino funcion para levantar resultados de enriquecimiento entre determinados niveles de jerarquia de GO
def go_enrichment_parser(xlsx_file):
    # saco tag del estadio
    estadio = xlsx_file.rpartition('/')[2].rpartition('_')[2].split('.')[0]
    # cargo la tabla
    tabla = pd.read_excel(xlsx_file).query("`enrichment` == 'e' and `NS` == 'BP' and `depth` >= 4 and `depth` <= 8")[['GO', 'p_fdr_bh']].assign(estadio = estadio)
    # saco descripcion
    tabla = tabla.assign(GO_descr = lambda df: [godag.get(x).name for x in df['GO'].to_list()])
    # pivoteo 
    tabla = tabla.pivot(index = ['GO', 'GO_descr'], values = 'p_fdr_bh', columns = 'estadio')
    # devuelvo
    return tabla.reset_index()
```

```python
estadio = xlsx_file.rpartition('/')[2].rpartition('_')[2].split('.')[0]
xlsx_file = '../go_enrichment_analysis/results/highly_expressed_genes/goatools/goea_results_sig_HE_IS21.xlsx'
tabla = pd.read_excel(xlsx_file).query("`enrichment` == 'e' and `NS` == 'BP'")[['GO', 'p_fdr_bh']].assign(estadio = estadio)
```

```python
is21 = go_enrichment_parser('../go_enrichment_analysis/results/highly_expressed_genes/goatools/goea_results_sig_HE_IS21.xlsx')
is15 = go_enrichment_parser('../go_enrichment_analysis/results/highly_expressed_genes/goatools/goea_results_sig_HE_IS15.xlsx')
is30 = go_enrichment_parser('../go_enrichment_analysis/results/highly_expressed_genes/goatools/goea_results_sig_HE_IS30.xlsx')
```

```python
import glob
gos_enriquecidos = [x for x in map(go_enrichment_parser, glob.glob('../go_enrichment_analysis/results/highly_expressed_genes/goatools/*.xlsx'))]
```

Saco cuales son los enriquecidos en MIR e IS (indices 3, 6, 8, 10)

```python
goids_enriquecidos_mir_is = list(set(pd.concat([gos_enriquecidos[3], gos_enriquecidos[6],
           gos_enriquecidos[8], gos_enriquecidos[10]])['GO'].to_list()))
```

```python
# se une tabla de GO terms de izquierda a derecha
from functools import reduce
df_merged = reduce(lambda  left,right: pd.merge(left,right, on=['GO', 'GO_descr'], how = 'outer'), gos_enriquecidos)
```

```python
df_merged
```

```python
# retengo para observar solo los GOs enriquecidos en estadios larvarios de F. hepatica
df_merged = df_merged.query("`GO` in @goids_enriquecidos_mir_is")
```

```python
# calculo clustering para los terminos involucrados
goids = df_merged['GO'].to_list()
```

```python
# sigo el protocolo que tiene la gente de GOATOOLS, calculando el semantic similarity de Wang
from goatools.semsim.termwise.wang import SsWang
relationships = {'part_of'}

# goids: researcher-provided GO terms
wang_r1 = SsWang(goids, godag, relationships)
```

```python
relationship2weight = {
    'is_a': 0.9,
    'part_of': 0.9
}
wang_r1 = SsWang(goids, godag, relationships, relationship2weight)
```

```python
# las distancias usadas son 1-indice de similitud semantica de Wang
distancias = []
for goid_a in goids:
    distancias.append([1-wang_r1.get_sim(goid_a, goid_b) for goid_b in goids])
```

```python
array_distancias = np.array(distancias)
```

```python
# ordeno la tabla de ploteo en base a ese clustering
df_merged = df_merged.rename({'mir': 'MIR'}, axis = 1)[['GO', 'GO_descr', 'EGG', 'MIR', 'IS15', 'IS21', 'IS30',
           'MET', 'NEJ1', 'NEJ3', 'NEJ24', 'JUV', 'AD']]
```

```python
import seaborn as sns
import scipy.cluster.hierarchy as hc

linkage = hc.linkage(array_distancias, method = 'ward')

para_plot = (df_merged.
             set_index(['GO_descr']))

para_plot['GO'] = pd.Categorical(para_plot['GO'].to_list(), goids)
para_plot = para_plot.sort_values(by = 'GO')

g = sns.clustermap(para_plot.reset_index().query("`GO` in @goids_enriquecidos_mir_is").drop('GO', axis = 1).set_index('GO_descr').applymap(lambda x: -np.log10(x)), 
               yticklabels = 1, cmap = 'Reds', vmin = 0, vmax = 10, row_linkage = linkage,
               method = 'ward', col_cluster = False).ax_heatmap
```

```python
g.set_yticklabels(g.get_ymajorticklabels(), fontsize = 2)
g.set_xticklabels(g.get_xmajorticklabels(), fontsize = 7)
g.set_xlabel('Estadío de desarrollo')
g.set_ylabel('Descripción de GO term')
```
